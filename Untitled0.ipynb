```python
import pandas as pd

# Load dataset
df = pd.read_csv("metadata_sample.csv")

# Show first 5 rows
df.head()
```

```python
# Shape of the dataset (rows, columns)
df.shape

# Column names
df.columns
```

```python
# Display first 5 rows
df.head()

```

```python
# Convert publish_time to datetime format
df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce')

# Extract the year into a new column
df['year'] = df['publish_time'].dt.year

# Create a new column: abstract word count
df['abstract_word_count'] = df['abstract'].astype(str).apply(lambda x: len(x.split()))

# Show a preview
df[['title', 'publish_time', 'year', 'abstract_word_count']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Number of papers published per year
papers_per_year = df['year'].value_counts().sort_index()

plt.figure(figsize=(8,5))
sns.barplot(x=papers_per_year.index, y=papers_per_year.values, color="skyblue")
plt.title("Number of Papers Published per Year")
plt.xlabel("Year")
plt.ylabel("Number of Papers")
plt.show()
```

```python
# Top 10 journals
top_journals = df['journal'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(y=top_journals.index, x=top_journals.values, palette="viridis")
plt.title("Top 10 Journals Publishing COVID-19 Research")
plt.xlabel("Number of Papers")
plt.ylabel("Journal")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['abstract_word_count'].dropna(), bins=50, kde=True, color="purple")
plt.title("Distribution of Abstract Word Counts")
plt.xlabel("Word Count")
plt.ylabel("Frequency")
plt.show()
```
